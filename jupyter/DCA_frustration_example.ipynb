```python
import dca_frustratometer
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
distance_matrix=dca_frustratometer.get_distance_matrix_from_pdb('../examples/data/1l63.pdb','A')
seq=dca_frustratometer.get_protein_sequence_from_pdb('../examples/data/1l63.pdb','A')
potts_model=dca_frustratometer.load_potts_model('../examples/data/PottsModel1l63A.mat')
```

```python
native_energy=dca_frustratometer.compute_native_energy(seq,potts_model,distance_matrix)
native_energy
```

```python
decoy_energy=dca_frustratometer.compute_singleresidue_decoy_energy(seq,potts_model,distance_matrix)
sns.clustermap(decoy_energy)
```

```python
frustration=dca_frustratometer.compute_singleresidue_frustration(decoy_energy, native_energy)
plt.plot(frustration)
```
